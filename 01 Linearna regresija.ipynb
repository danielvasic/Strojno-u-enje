## Analiza regresije

Analiza regresije traži odnose među varijablama. Na primjer, možete promatrati nekoliko zaposlenika neke tvrtke i pokušati razumjeti kako njihove plaće ovise o njihovim značajkama, kao što su iskustvo, razina obrazovanja, uloga, grad zaposlenja i slično. 

Ovo je problem regresije gdje podaci povezani s svakim zaposlenikom predstavljaju jedno opažanje. Pretpostavka je da su iskustvo, obrazovanje, uloga i grad neovisne značajke, dok plaća ovisi o njima. Slično tome, možete pokušati uspostaviti matematičku ovisnost cijena kuća o površini, broju spavaćih soba, udaljenosti od centra grada i tako dalje. 

Općenito, u analizi regresije, razmatrate neki fenomen od interesa i imate niz opažanja. Svako opažanje ima dvije ili više značajki. Slijedeći pretpostavku da barem jedna od značajki ovisi o ostalima, pokušavate uspostaviti odnos među njima. Drugim riječima, potrebno je pronaći funkciju koja dovoljno dobro mapira neke značajke ili varijable na druge. 

Ovisne značajke nazivaju se **ovisnim varijablama**, **izlazima** ili **odgovorima**. Neovisne značajke nazivaju se neovisnim varijablama, **ulazima**, **regresorima** ili **prediktorima**. Problemi regresije obično imaju jednu kontinuiranu i neograničenu ovisnu varijablu. Ulazi, međutim, mogu biti kontinuirani, diskretni ili čak kategorički podaci kao što su spol, nacionalnost ili marka.

Uobičajena je praksa označavanje izlaza s 𝑦 i ulaza s 𝑥. Ako postoje dvije ili više neovisnih varijabli, onda se mogu prikazati kao vektor 𝐱 = (𝑥₁, …, 𝑥ᵣ), gdje je 𝑟 broj ulaza.


## Linearna regresija

Linearna regresija je vjerojatno jedna od najvažnijih i najšire korištenih tehnika regresije. To je jedna od najjednostavnijih metoda regresije. Jedna od glavnih prednosti je lakoća interpretacije rezultata.

### Formulacija problema

Kada implementirate linearnu regresiju neke ovisne varijable 𝑦 na skupu neovisnih varijabli 𝐱 = (𝑥₁, …, 𝑥ᵣ), gdje je 𝑟 broj prediktora, pretpostavljate linearni odnos između 𝑦 i 𝐱: 𝑦 = 𝛽₀ + 𝛽₁𝑥₁ + ⋯ + 𝛽ᵣ𝑥ᵣ + 𝜀. Ova jednadžba je jednadžba regresije. 𝛽₀, 𝛽₁, …, 𝛽ᵣ su koeficijenti regresije, a 𝜀 je slučajna greška.

Linearna regresija izračunava procjenitelje koeficijenata regresije ili jednostavno predviđene težine, označene sa 𝑏₀, 𝑏₁, …, 𝑏ᵣ. Ovi procjenitelji definiraju procijenjenu regresijsku funkciju 𝑓(𝐱) = 𝑏₀ + 𝑏₁𝑥₁ + ⋯ + 𝑏ᵣ𝑥ᵣ. Ova funkcija bi trebala dovoljno dobro uhvatiti ovisnosti između ulaza i izlaza.

Procijenjeni ili predviđeni odgovor, 𝑓(𝐱ᵢ), za svako opažanje 𝑖 = 1, …, 𝑛, trebao bi biti što bliži odgovarajućem stvarnom odgovoru 𝑦ᵢ. Razlike 𝑦ᵢ - 𝑓(𝐱ᵢ) za sva opažanja 𝑖 = 1, …, 𝑛, nazivaju se reziduali. Regresija je o određivanju najboljih predviđenih težina - to jest, težina koje odgovaraju najmanjim rezidualima.

Da biste dobili najbolje težine, obično minimizirate sumu kvadriranih reziduala (SSR) za sva opažanja 𝑖 = 1, …, 𝑛: SSR = Σᵢ(𝑦ᵢ - 𝑓(𝐱ᵢ))². Ovaj pristup naziva se metoda običnih najmanjih kvadrata.

### Učinkovitost regresije

Varijacija stvarnih odgovora 𝑦ᵢ, 𝑖 = 1, …, 𝑛, djelomično se događa zbog ovisnosti o prediktorima 𝐱ᵢ. Međutim, postoji i dodatna inherentna varijacija izlaza.

Koeficijent određenosti, označen kao 𝑅², govori vam koja količina varijacije u 𝑦 može biti objašnjena ovisnošću o 𝐱, koristeći određeni regresijski model. Veći 𝑅² ukazuje na bolje usklađivanje i znači da model može bolje objasniti varijaciju izlaza s različitim ulazima.

Vrijednost 𝑅² = 1 odgovara SSR = 0. To je savršeno usklađivanje, budući da se vrijednosti predviđenih i stvarnih odgovora potpuno podudaraju.


# Jednostavna linearna regresija

Jednostavna ili univarijatna linearna regresija je najjednostavniji slučaj linearne regresije, jer ima samo jednu neovisnu varijablu, 𝐱 = 𝑥.

Kada implementirate jednostavnu linearnu regresiju, obično započinjete s danim skupom ulazno-izlaznih parova (𝑥-𝑦). Ti parovi su vaša opažanja, prikazana kao zelene kružnice na slici. Na primjer, najljevija opažanja imaju ulaz 𝑥 = 5 i stvarni izlaz, ili odgovor, 𝑦 = 5. Sljedeća ima 𝑥 = 15 i 𝑦 = 20, i tako dalje.

Procijenjena regresijska funkcija, predstavljena crnom linijom, ima jednadžbu 𝑓(𝑥) = 𝑏₀ + 𝑏₁𝑥. Vaš cilj je izračunati optimalne vrijednosti predviđenih težina 𝑏₀ i 𝑏₁ koje minimiziraju SSR i određuju procijenjenu regresijsku funkciju.

Vrijednost 𝑏₀, također nazvana presjek, pokazuje točku gdje procijenjena regresijska linija prelazi os 𝑦. To je vrijednost procijenjenog odgovora 𝑓(𝑥) za 𝑥 = 0. Vrijednost 𝑏₁ određuje nagib procijenjene regresijske linije.

Predviđeni odgovori, prikazani kao crveni kvadrati, su točke na regresijskoj liniji koje odgovaraju ulaznim vrijednostima. Na primjer, za ulaz 𝑥 = 5, predviđeni odgovor je 𝑓(5) = 8.33, što predstavlja najljeviji crveni kvadrat.

Vertikalne isprekidane sive linije predstavljaju reziduale, koje se mogu izračunati kao 𝑦ᵢ - 𝑓(𝐱ᵢ) = 𝑦ᵢ - 𝑏₀ - 𝑏₁𝑥ᵢ za 𝑖 = 1, …, 𝑛. To su udaljenosti između zelenih kružnica i crvenih kvadrata. Kada implementirate linearnu regresiju, zapravo pokušavate minimizirati te udaljenosti i učiniti crvene kvadrate što bližima unaprijed definiranim zelenim kružnicama.


```python
import matplotlib.pyplot as plt
import numpy as np

# Definiranje ulaznih i izlaznih vrijednosti
x = np.array([5, 15, 25, 35, 45, 55])
y = np.array([5, 20, 14, 32, 22, 38])

# Izračunavanje koeficijenata linearne regresije
b1, b0 = np.polyfit(x, y, 1)

# Izračunavanje predviđenih odgovora
y_pred = b0 + b1 * x

# Plotanje regresijske linije
plt.plot(x, y_pred, color='black', label='Regresijska linija')

# Plotanje opažanja kao zelenih kružnica
plt.scatter(x, y, color='green', label='Opažanja')

# Plotanje predviđenih odgovora kao crvenih kvadrata
plt.scatter(x, y_pred, color='red', label='Predviđeni odgovori')



# Dodavanje reziduala kao sivih isprekidanih linija
for xi, yi, ypi in zip(x, y, y_pred):
    plt.plot([xi, xi], [yi, ypi], color='gray', linestyle='dashed')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Primjer jednostavne linearne regresije')
plt.legend()
plt.show()
```
