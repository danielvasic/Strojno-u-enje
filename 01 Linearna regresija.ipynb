## Analiza regresije

Analiza regresije tra≈æi odnose meƒëu varijablama. Na primjer, mo≈æete promatrati nekoliko zaposlenika neke tvrtke i poku≈°ati razumjeti kako njihove plaƒáe ovise o njihovim znaƒçajkama, kao ≈°to su iskustvo, razina obrazovanja, uloga, grad zaposlenja i sliƒçno. 

Ovo je problem regresije gdje podaci povezani s svakim zaposlenikom predstavljaju jedno opa≈æanje. Pretpostavka je da su iskustvo, obrazovanje, uloga i grad neovisne znaƒçajke, dok plaƒáa ovisi o njima. Sliƒçno tome, mo≈æete poku≈°ati uspostaviti matematiƒçku ovisnost cijena kuƒáa o povr≈°ini, broju spavaƒáih soba, udaljenosti od centra grada i tako dalje. 

Opƒáenito, u analizi regresije, razmatrate neki fenomen od interesa i imate niz opa≈æanja. Svako opa≈æanje ima dvije ili vi≈°e znaƒçajki. Slijedeƒái pretpostavku da barem jedna od znaƒçajki ovisi o ostalima, poku≈°avate uspostaviti odnos meƒëu njima. Drugim rijeƒçima, potrebno je pronaƒái funkciju koja dovoljno dobro mapira neke znaƒçajke ili varijable na druge. 

Ovisne znaƒçajke nazivaju se **ovisnim varijablama**, **izlazima** ili **odgovorima**. Neovisne znaƒçajke nazivaju se neovisnim varijablama, **ulazima**, **regresorima** ili **prediktorima**. Problemi regresije obiƒçno imaju jednu kontinuiranu i neograniƒçenu ovisnu varijablu. Ulazi, meƒëutim, mogu biti kontinuirani, diskretni ili ƒçak kategoriƒçki podaci kao ≈°to su spol, nacionalnost ili marka.

Uobiƒçajena je praksa oznaƒçavanje izlaza s ùë¶ i ulaza s ùë•. Ako postoje dvije ili vi≈°e neovisnih varijabli, onda se mogu prikazati kao vektor ùê± = (ùë•‚ÇÅ, ‚Ä¶, ùë•·µ£), gdje je ùëü broj ulaza.


## Linearna regresija

Linearna regresija je vjerojatno jedna od najva≈ænijih i naj≈°ire kori≈°tenih tehnika regresije. To je jedna od najjednostavnijih metoda regresije. Jedna od glavnih prednosti je lakoƒáa interpretacije rezultata.

### Formulacija problema

Kada implementirate linearnu regresiju neke ovisne varijable ùë¶ na skupu neovisnih varijabli ùê± = (ùë•‚ÇÅ, ‚Ä¶, ùë•·µ£), gdje je ùëü broj prediktora, pretpostavljate linearni odnos izmeƒëu ùë¶ i ùê±: ùë¶ = ùõΩ‚ÇÄ + ùõΩ‚ÇÅùë•‚ÇÅ + ‚ãØ + ùõΩ·µ£ùë•·µ£ + ùúÄ. Ova jednad≈æba je jednad≈æba regresije. ùõΩ‚ÇÄ, ùõΩ‚ÇÅ, ‚Ä¶, ùõΩ·µ£ su koeficijenti regresije, a ùúÄ je sluƒçajna gre≈°ka.

Linearna regresija izraƒçunava procjenitelje koeficijenata regresije ili jednostavno predviƒëene te≈æine, oznaƒçene sa ùëè‚ÇÄ, ùëè‚ÇÅ, ‚Ä¶, ùëè·µ£. Ovi procjenitelji definiraju procijenjenu regresijsku funkciju ùëì(ùê±) = ùëè‚ÇÄ + ùëè‚ÇÅùë•‚ÇÅ + ‚ãØ + ùëè·µ£ùë•·µ£. Ova funkcija bi trebala dovoljno dobro uhvatiti ovisnosti izmeƒëu ulaza i izlaza.

Procijenjeni ili predviƒëeni odgovor, ùëì(ùê±·µ¢), za svako opa≈æanje ùëñ = 1, ‚Ä¶, ùëõ, trebao bi biti ≈°to bli≈æi odgovarajuƒáem stvarnom odgovoru ùë¶·µ¢. Razlike ùë¶·µ¢ - ùëì(ùê±·µ¢) za sva opa≈æanja ùëñ = 1, ‚Ä¶, ùëõ, nazivaju se reziduali. Regresija je o odreƒëivanju najboljih predviƒëenih te≈æina - to jest, te≈æina koje odgovaraju najmanjim rezidualima.

Da biste dobili najbolje te≈æine, obiƒçno minimizirate sumu kvadriranih reziduala (SSR) za sva opa≈æanja ùëñ = 1, ‚Ä¶, ùëõ: SSR = Œ£·µ¢(ùë¶·µ¢ - ùëì(ùê±·µ¢))¬≤. Ovaj pristup naziva se metoda obiƒçnih najmanjih kvadrata.

### Uƒçinkovitost regresije

Varijacija stvarnih odgovora ùë¶·µ¢, ùëñ = 1, ‚Ä¶, ùëõ, djelomiƒçno se dogaƒëa zbog ovisnosti o prediktorima ùê±·µ¢. Meƒëutim, postoji i dodatna inherentna varijacija izlaza.

Koeficijent odreƒëenosti, oznaƒçen kao ùëÖ¬≤, govori vam koja koliƒçina varijacije u ùë¶ mo≈æe biti obja≈°njena ovisno≈°ƒáu o ùê±, koristeƒái odreƒëeni regresijski model. Veƒái ùëÖ¬≤ ukazuje na bolje usklaƒëivanje i znaƒçi da model mo≈æe bolje objasniti varijaciju izlaza s razliƒçitim ulazima.

Vrijednost ùëÖ¬≤ = 1 odgovara SSR = 0. To je savr≈°eno usklaƒëivanje, buduƒái da se vrijednosti predviƒëenih i stvarnih odgovora potpuno podudaraju.


# Jednostavna linearna regresija

Jednostavna ili univarijatna linearna regresija je najjednostavniji sluƒçaj linearne regresije, jer ima samo jednu neovisnu varijablu, ùê± = ùë•.

Kada implementirate jednostavnu linearnu regresiju, obiƒçno zapoƒçinjete s danim skupom ulazno-izlaznih parova (ùë•-ùë¶). Ti parovi su va≈°a opa≈æanja, prikazana kao zelene kru≈ænice na slici. Na primjer, najljevija opa≈æanja imaju ulaz ùë• = 5 i stvarni izlaz, ili odgovor, ùë¶ = 5. Sljedeƒáa ima ùë• = 15 i ùë¶ = 20, i tako dalje.

Procijenjena regresijska funkcija, predstavljena crnom linijom, ima jednad≈æbu ùëì(ùë•) = ùëè‚ÇÄ + ùëè‚ÇÅùë•. Va≈° cilj je izraƒçunati optimalne vrijednosti predviƒëenih te≈æina ùëè‚ÇÄ i ùëè‚ÇÅ koje minimiziraju SSR i odreƒëuju procijenjenu regresijsku funkciju.

Vrijednost ùëè‚ÇÄ, takoƒëer nazvana presjek, pokazuje toƒçku gdje procijenjena regresijska linija prelazi os ùë¶. To je vrijednost procijenjenog odgovora ùëì(ùë•) za ùë• = 0. Vrijednost ùëè‚ÇÅ odreƒëuje nagib procijenjene regresijske linije.

Predviƒëeni odgovori, prikazani kao crveni kvadrati, su toƒçke na regresijskoj liniji koje odgovaraju ulaznim vrijednostima. Na primjer, za ulaz ùë• = 5, predviƒëeni odgovor je ùëì(5) = 8.33, ≈°to predstavlja najljeviji crveni kvadrat.

Vertikalne isprekidane sive linije predstavljaju reziduale, koje se mogu izraƒçunati kao ùë¶·µ¢ - ùëì(ùê±·µ¢) = ùë¶·µ¢ - ùëè‚ÇÄ - ùëè‚ÇÅùë•·µ¢ za ùëñ = 1, ‚Ä¶, ùëõ. To su udaljenosti izmeƒëu zelenih kru≈ænica i crvenih kvadrata. Kada implementirate linearnu regresiju, zapravo poku≈°avate minimizirati te udaljenosti i uƒçiniti crvene kvadrate ≈°to bli≈æima unaprijed definiranim zelenim kru≈ænicama.


```python
import matplotlib.pyplot as plt
import numpy as np

# Definiranje ulaznih i izlaznih vrijednosti
x = np.array([5, 15, 25, 35, 45, 55])
y = np.array([5, 20, 14, 32, 22, 38])

# Izraƒçunavanje koeficijenata linearne regresije
b1, b0 = np.polyfit(x, y, 1)

# Izraƒçunavanje predviƒëenih odgovora
y_pred = b0 + b1 * x

# Plotanje regresijske linije
plt.plot(x, y_pred, color='black', label='Regresijska linija')

# Plotanje opa≈æanja kao zelenih kru≈ænica
plt.scatter(x, y, color='green', label='Opa≈æanja')

# Plotanje predviƒëenih odgovora kao crvenih kvadrata
plt.scatter(x, y_pred, color='red', label='Predviƒëeni odgovori')



# Dodavanje reziduala kao sivih isprekidanih linija
for xi, yi, ypi in zip(x, y, y_pred):
    plt.plot([xi, xi], [yi, ypi], color='gray', linestyle='dashed')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Primjer jednostavne linearne regresije')
plt.legend()
plt.show()
```
